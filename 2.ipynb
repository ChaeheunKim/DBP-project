```python
import os
import shutil

def process_files(input_folder, output_folder, extensions):
    # 파일 경로를 저장할 리스트 초기화
    file_paths = []

    # 현재 폴더 및 하위 폴더에서 파일 찾기
    for root, _, files in os.walk(input_folder):
        for file in files:
            if any(file.lower().endswith(ext) for ext in extensions):
                file_path = os.path.join(root, file)
                file_paths.append(file_path)

    # 파일 개수 확인
    total_files = len(file_paths)
    print("file count: ", total_files)
    processed_files = 0

    # 파일을 지정한 폴더로 복사
    for file_path in file_paths:
        file_name = os.path.basename(file_path)
        shutil.copy(file_path, os.path.join(output_folder, file_name))
        processed_files += 1

        # 처리 상황 출력
        print(f"DATACONVERTER [{input_folder}->{output_folder}] 진행 상황: {processed_files}/{total_files} 파일 복사 완료")

# 이미지와 레이블 폴더 및 결과를 저장할 폴더 경로 설정
image_folder = r'age_val/image1'
label_folder = r'age_val/label1'
output_image_folder = r'age_val_result/image'
output_label_folder = r'age_val_result/label'

# 이미지 및 레이블 처리 함수 호출
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']  # 이미지 파일 확장자
label_extensions = ['.json']  # 레이블 파일 확장자

process_files(image_folder, output_image_folder, image_extensions)  # 이미지 처리
process_files(label_folder, output_label_folder, label_extensions)  # 레이블 처리
```
